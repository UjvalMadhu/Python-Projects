BlackJack is a classic card game. It is designed here to be played against the computer as the dealer.
The goal of the game is to not get to go above a sum of 21 in hand.
The rules of the game are as follows:
1. From a shuffled deck of cards 2 cards are handed to the player and the dealer.
2. The player's cards are displayed but the dealers last card is kept hidden.
3. Next the player can choose to 'Hit' meaning to get a card from the deck or 'Stand' meaning to skip getting a card from the deck. The player and dealer go through each round alternatively, but can only 'Stand' for three rounds in a row and will have to 'Hit' in the fourth round in this case.
4. The first to not get a value above 21 in hand wins the game and the bet.
5. You can choose to buy chips to place bets and track your chips in this game.

### Card Values
There are four suits in a deck- "Hearts',"Diamonds","Spades" and "Clubs" having 13 cards in each suit ,constituting of 52 total cards.
There are 9 numbered cards and 4 face cards in each suit.
The numbered cards (2 to 10) of each of the four suits have the same value per card as their respective number indicates.
The face cards 'King', 'Queen' and 'Jack' have vales of 10 each. and 'Ace' cards can have a value of 11 or 1, which is automatically adjusted if total value goes above 21.

#### Step1: 
Here we import the random module so that we can use the random function to shuffle our cards. We also define global variables which can be accessed thoughout the code to define our deck of cards.

```python
import random                                      # Imports the random module

suits = ['Hearts','Diamonds','Spades','Clubs']     # List Defining the four Suits

ranks = ['Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','King','Queen','Ace'] # List defining the...  
                                                                                                     # ...ranks of cards
values ={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'King':10,'Queen':10,
         'Ace':11}                                  # Dictionary matching each rank to its numerical value used for calculation
```

#### Step 2:
The Card class is used to define a card. 

```python
class Card:
    
    def __init__(self,suit,rank):             # The suit and rank are the two properties of a card.
        self.suit = suit
        self.rank = rank
        self.value = values[rank]             # The numerical value of the card is calculated from its rank.
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

#### Step 3:
The Deck class is used to define the deck of cards used in the game.

```python
class Deck:
    
    def __init__(self):
        self.all_cards = []                                  # The all_cards list is used to hold all 52 the cards 
        for suit in suits:                                   # The for loops creates and adds all the card objects to aal_cards 
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
    
    def shuffle(self):                                       # The shuffle method is used for shuffling the cards in the deck
        random.shuffle(self.all_cards)
        
    def deal_one(self):                                     # The deal_one function deals the top card from the deck when called
        return self.all_cards.pop()
```

#### Step 4:
The Player class defines the fuctions that does all operations regarding the chips and bets of the player.

```python
class Player:
    
    def __init__(self,name,chips):
        self.name =name
        self.chips = chips
        
    def __str__(self):
        return f"Name: {self.name} \nChips: {self.chips}"   # Displayes the name and current chips of the player
    
    def buy_chips(self):                                    # Function that lets the player buy chips before the game
        buy = ""
        while True:
            try:                                            # The try statement works when the input is either Y or N 
                buy = str(input("\nDo you wish to buy chips? Y/N :").upper())    # The input is always transformed to uppercase
                if buy == "Y":
                    while True:                             # Here the try staement works when the input is a positive integer
                        try:
                            p = int(input("Enter the value of chips to Buy:"))
                            self.chips += p
                            print(self)
                        except:
                            print("Please enter an integer Value")
                        else:
                            break
                elif buy == "N":
                        print(self)
                else:
                    print("Please enter a valid answer.")
                    continue
            except:
                 print("please enter a Valid Option")
            else:
                break
                
    
    def place_bet(self):                                     #This function lets the player set a bet and checks its validity
        bet = 0                                              # The bet can never be more than the chips the player has
        while True:
            try:
                bet = int(input("Enter the value of bet $:"))

            except:
                print("Please enter an Integer Value.")
                continue

            else:
                if bet< self.chips or bet == self.chips:
                    self.chips -= bet
                    print(f'You have placed a bet of {bet}$ and You now have chips worth {self.chips}$ remaining.\n')
                    return bet
                    break
                else:
                    print("You do not have sufficient chips for this bet.")
                    while True:
                        buy = str(input("You need to buy more chips for this bet, proceed? Y/N")).lower()
                        if buy != "y" and buy !="n":
                            print("Please enter a valid respone.")
                            continue
                        else:
                            break
                    if buy == "y":
                        self.buy_chips()
                        continue
                    elif buy == "n" :
                        print("Please set a lower bet.")
                        continue
                break
               
    
    def win_bet(self,bet):                                         # This function modifies chips when the player wins 
        self.bet = bet
        self.chips += (2*self.bet) 
        print(f"\n\n{self.name} has won the bet of {self.bet}!!")
        print(self)
              
    def gave_bet(self,bet):                                        # This function modifies the chips when the computer wins
        self.bet = bet
        print(f"\n\n{self.name} has gave the bet of {self.bet}!!")
        print(self)
        
    def draw_bet(self,bet):                                        # This function executes when the game is a draw
        self.bet = bet
        self.chips += self.bet
        print(f"\n\nThe Game was a draw!!")
        print(self)
```

#### Step 5:
The GameInit Class defines some important functions that helps run the game.

```python
class GameInit:
    
    def __init__(self):                                     # These variables can be used by all functions in this class
        self.p1_cards = []                                  # p1_cards reprsents the cards in the player's hand.
        self.p2_cards = []                                  # p2_cards reprsents the cards in the dealer's hand.
        self.round = 1                                      # round represents the round of game
        self.value1 = 0                                     # value1 is the value of the player's hand.
        self.value2 = 0                                     # value2 is the value of the dealer's hand.
        self.len1 = 0                                       # len1 reprsents the number of cards in the player's hand
        self.len2 = 0                                       # len2 reprsents the number of cards in the dealer's hand
        self.ranks1 =[]                                     # ranks1 records the ranks of all the cards in the player's hand
        self.ranks2 =[]                                     # ranks2 records the ranks of all the cards in the Dealer's hand
        self.p1_ace =[]                                     # p1_ace, p2_ace records a 0 for every new card added and sets to..
        self.p2_ace =[]                                     #... 1 for an ace card that was adjusted to a value of 1 from 11.
        self.p_hold = 0                                     # p_hold and c_hold are used to count the number of stands in a row
        self.c_hold = 0
    
    
    def player_init(self,deck):                             # This function initiates the game by dealing 2 cards... 
                                                            #... to the player and the dealer and prints the player's cards
        for i in range(0,2):
            self.p1_cards.append(deck.deal_one())
            self.p2_cards.append(deck.deal_one())
        print(f"Both players have recieved 2 cards each.\n Your cards are:")
        for i in range (0,2):
            print(self.p1_cards[i])
            
            
    def ace_adjust(self,ranks,value,aces_check):            # This function adjusts the value of an ace card in hand ...
        self.ranks = ranks                                  #... to 1 if total hand value exceeds 21
        self.value = int(value)                             # aces_check records 1 for the adjusted ace's position
        self.aces_check = aces_check
        
        for i in range(0,len(self.ranks)):
            if self.ranks[i] == "Ace":
                if self.value > 21 and int(self.aces_check[i]) == 0:
                    self.value-=10
                    print("Value of ACE adjusted to 1.")
                    print(self.value)
                    self.aces_check[i] = 1
                    continue
                continue
        return self.value
                
            
    
    def set_table(self):                                   # This function sets up the first round(=1) of the game
                                                           # ... after each recieves their cards using player_init
        if self.round ==1:                                 
            self.len1 = len(self.p1_cards)                 # The number of cards is recorded so as to print them
            self.len2 = len(self.p2_cards)


            print("\n \nPlayer Hand: ",end=" ")                     # Player's Hand
            for i in range(0,self.len1):
                self.value1 += self.p1_cards[i].value               # The total value of all cards in hand is summed
                self.ranks1.append(self.p1_cards[i].rank)
                self.p1_ace.append(0)
                print(self.p1_cards[i],end="  :  ")       
                if i == self.len1-1:
                    self.value1 = int(self.ace_adjust(self.ranks1,self.value1,self.p1_ace)) # Adjusting the Ace values in hand 
                    print(f"player value = {self.value1}\n\n")


            print("Computer Hand: ",end=" ")                         # Dealer's Hand
            for i in range(0,self.len2):
                self.value2 +=self.p2_cards[i].value                 # The total value of all cards in hand is summed
                self.ranks2.append(self.p2_cards[i].rank)
                self.p2_ace.append(0)
                if i != self.len2-1:
                    print(self.p2_cards[i],end="  :  ")
                else:
                    self.value2 = int(self.ace_adjust(self.ranks2,self.value2,self.p2_ace)) # Adjusting the Ace values in hand
                    print(f"Hidden Card : Computer Value = Hidden\n\n")
                    
                    
    def players_hit(self,deck):                                  # This function defines how the game is played after the...
                                                                 # ... first round.
        if self.round>1:
           
            while True:
                hit = input("\nDo You want to Hit?: press'H' for hit and 'S' for stand: ")
                if (hit.lower() == "h") or (self.p_hold > 2):
                    if self.p_hold > 2:
                        print("\n You have run out of stand options!")
                    ("Player Chooses to Hit.")
                    self.p1_cards.append(deck.deal_one())
                    self.len1 += 1
                    self.value1 += self.p1_cards[self.len1-1].value          #Indexing starts at 0, so is 1 - card position
                    self.ranks1.append(self.p1_cards[self.len1-1].rank)
                    self.p1_ace.append(0)
                    self.value1 = int(self.ace_adjust(self.ranks1,self.value1,self.p1_ace)) 
                    self.p_hold = 0                                   # hold values are set to 0 after a hit.

                    
                elif hit.lower() == "s" and self.p_hold <= 2:
                    print("Player Chooses to Stand.\n")
                    self.p_hold += 1
                    print(f"You have '\033[1m{3- self.p_hold}\033[0m' stands remaining.") 
                    
               
                else:
                    print("Try Again")
                    continue
            
                print("\nComputer's Turn")                         # Here the computer plays the game 

                i = random.randint(0,1)                          # i, j are variables that have a 50% probability of getting...
                j = random.randint(0,1)                          # ... a 1 or 0 respectively

                if self.value2 <= 15 or (i == 1 and j==1) or self.c_hold > 3:  # There is a 25% probability that the computer will...
                    print("\nComputer decides to hit")                      # ...hit when the value in hand is already above 15

                    self.p2_cards.append(deck.deal_one())
                    self.len2 += 1
                    self.value2 += self.p2_cards[self.len2-1].value
                    self.ranks2.append(self.p2_cards[self.len2-1].rank)
                    self.p2_ace.append(0)
                    self.value2 = int(self.ace_adjust(self.ranks2,self.value2,self.p2_ace))
                    self.c_hold = 0                                  # hold values are set to 0 after a hit.                                  

                else:
                    print("\nComputer decides to stand.")
                    self.c_hold +=1
                    print(f"Computer has \033[1m{3-self.c_hold}\033[0m stands remaining.") 
                
                print("\nPlayer Hand: ",end=" ")
                for i in range(0,self.len1):
                    print(self.p1_cards[i],end="  :  ")
                print(f"player value = {self.value1}")
                    
                print("Computer Hand: ",end=" ")
                for i in range(0,self.len2):
                    print(self.p2_cards[i],end="  :  ")
                print(f"Computer value = {self.value2}")
                break
```
